```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
## Line graph over time

df = pd.read_csv('/Users/angelikaarada/Documents/test.csv')

df.head()
```

```python
plt.figure(figsize=(8,5))

plt.title('Resolution of Crystal or CryoEM Structures over Time', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.plot(df.Year, df.Res, 'g.-', label='Resolution')

# Another Way to plot many values!
# countries_to_look_at = ['Australia', 'USA', 'Canada', 'South Korea']
# for country in gas:
#     if country in countries_to_look_at:
#         plt.plot(gas.Year, gas[country], marker='.')

# plt.xticks(gas.Year[::3].tolist()+[2011])

plt.xlabel('Year')
plt.ylabel('Resolution')

plt.legend()

plt.savefig('R.png', dpi=300)

plt.show()
```
